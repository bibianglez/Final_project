```python
import pandas as pd
import eurostat
```

```python
table_code = 'demo_r_mlifexp'
df_life_exp = eurostat.get_data_df(table_code)
display(df_life_exp.head())
```

```python
df_life_exp.rename(columns={col : col.strip() for col in df_life_exp.columns}, inplace = True)
df_life_exp.rename(columns={col : col.replace("\\", "_").lower() for col in df_life_exp.columns}, inplace = True)
df_life_exp.shape
```

```python
df_life_exp.dtypes
```

```python
df_life_exp_clean = df_life_exp.drop(columns=['1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009'])
df_life_exp_clean.head()
```

```python
df_life_exp_clean.geo_time_period.unique()
```

```python
df_life_exp_nuts2 = df_life_exp_clean[df_life_exp_clean['geo_time_period'].str.len() == 4]
df_life_exp_nuts2.geo_time_period.unique()
```

```python
def df_info(df):
    print("\n    ----- Shape -----")
    print(df.shape)
    print("\n    ---- df info ---")
    print(df.info())
    print("\n    ---- Empty Values ---")
    print(df.isna().sum())
    print("\n    ----- Dtypes -----")
    print(df.dtypes)
    print("\n    ---- Statistics ---")
    return df.describe()
df_info(df_life_exp_nuts2)
```
